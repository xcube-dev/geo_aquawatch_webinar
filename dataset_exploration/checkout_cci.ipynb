```python
import re

from xcube.core.store import new_data_store
```

```python
store = new_data_store("cciodp")
```

```python
all_data_ids = list(store.get_data_ids())
len(all_data_ids)
```

```python
data_ids_sel = [d for d in all_data_ids if re.search(".*SST.*day.*", d)]
data_ids_sel
```

```python
for data_id in data_ids_sel:
    ds = store.open_data(data_id)
    print(ds.time[-1].values)
```

```python
ds.analysed_sst.isel(time=-1).sel(lat=slice(40, 60), lon=slice(-10, 20)).plot()
```
