## Setup the usage of xcube Viewer

The xcube Viewer is part of the xcube package which can be installed using either `pip` or `conda/mamba` from the conda-forge channel.

- üì¶ **PyPI:** [xcube-core on PyPI](https://pypi.org/project/xcube-core/)  `pip install xcube-core`
- üêç **Conda (conda-forge):** [xcube on Anaconda](https://anaconda.org/conda-forge/xcube) `conda install -c conda-forge xcube`

  You can also use **Mamba** as a faster alternative to Conda: `mamba install -c conda-forge xcube`
  
---

````python
from xcube.webapi.viewer import Viewer
from xcube.util.config import load_configs
````

## Option 1

Open xcube Viewer for a dataset instances opened or otherwise created in the same notebook.

```python
viewer = Viewer()

viewer.add_dataset(ds1)
viewer.add_dataset(ds2)

viewer.info()
viewer.show()
```

## Option 2

Use custom server configuration to start server and pass it to the viewer constructor. In this example, we have create a local file with the configuration and pass it to the Viewer.

````python
viewer = Viewer(server_config=load_configs("server-config.yaml"))
````

````python
viewer.info()
````

````python
viewer.show()
````

````python
viewer.stop_server()
````
